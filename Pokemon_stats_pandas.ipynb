# Load data into pandas 

```python
import pandas as pd

pok = pd.read_csv('Pokemon.csv') 

#pok.head() #shows top 5 rows
pok.tail() # shows bottom 5 rows
```

# Reading data in Pandas

```python
# reading columns/header
print(pok.columns)
```

```python
#reading columns
print(pok['Name'])
```

```python
#printing each row
print(pok.iloc[1]) #row with index 1 is given below
```

```python
#printing data from specific location
print(pok.iloc[2,3])
```

```python
#loc : can be also used with boolean values. Example below
pok.loc[(pok['Type 1'] == 'Grass') & (pok['Type 2'] == 'Fighting')]
```

```python
#Describing data

pok.describe()
```

```python
#sorting values
#pok.sort_values('Name', ascending = False)
pok.sort_values(['Type 1','HP'], ascending = [1,0])
```

# Making changes in data

```python
pok['Total'] = pok.iloc[:, 4:10].sum(axis=1)

cols = list(pok.columns)

pok = pok[cols[0:4] + [cols[-1]] + cols[4:12]]
```

### Exporting Data into desired format

```python
#pok.to_csv('modified.csv', index=False)

#pok.to_excel('modifiedtoexcel.xlsx', index = False) 

pok.to_csv('modified.txt', index= False, sep = '\t')
```

### Filtering the data

```python
new_df = pok.loc[(pok['Type 1'] == 'Fire' ) & (pok['Type 2'] == 'Flying' ) & (pok['HP'] > 70) ]

#pok.loc[(pok['Type 1'] == 'Fire' ) | (pok['Type 2'] == 'Flying' ) ]

new_df = new_df.reset_index(drop=True) #this reset the index shown in first column
new_df
```

```python
#filtering out data where name contains Mega word

#pok.loc[pok['Name'].str.contains('Mega')]
pok.loc[~pok['Name'].str.contains('Mega')]

#the second line drops all those rows where name contains Mega word. use of ~. reversed. 
```

```python
#using regular expressions
#re.I is ignoring the case

import re

#pok.loc[pok['Type 1'].str.contains('fire|grass', flags = re.I ,regex = True)]

pok.loc[pok['Name'].str.contains('^pi[a-z]*', flags = re.I ,regex = True)] #giving us names starting with pi 
```

### Conditional Changes

```python
#changing the calues in a row where type == flame

#pok.loc[pok['Type 1'] == 'Fire', 'Type 1' ] = 'Flamer'
pok.loc[pok['Type 1'] == 'Flamer', 'Type 1' ] = 'Fire'

pok
```

```python
#making all fire pokemon legendary

pok.loc[pok['Type 1'] == 'Fire', 'Legendary' ] = True

pok
```

### Aggregate statistics. Using groupby

```python
pok = pd.read_csv('modified.csv')

pok.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
```

```python
pok['Count'] = 1

pok.groupby(['Type 1', 'Type 2']).count()['Count'].sort_values(ascending = False)
```

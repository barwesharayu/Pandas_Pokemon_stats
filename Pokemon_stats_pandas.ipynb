# Load data into pandas 

```python
import pandas as pd

pok = pd.read_csv('Pokemon.csv') 

#pok.head() #shows top 5 rows
pok.tail() # shows bottom 5 rows
```

# Reading data in Pandas

```python
# reading columns/header
print(pok.columns)
```

```python
#reading columns
print(pok['Name'])
```

```python
#printing each row
print(pok.iloc[1]) #row with index 1 is given below
```

```python
#printing data from specific location
print(pok.iloc[2,3])
```

```python
#loc : can be also used with boolean values. Example below
pok.loc[(pok['Type 1'] == 'Grass') & (pok['Type 2'] == 'Fighting')]
```

```python
#Describing data

pok.describe()
```

```python
#sorting values
#pok.sort_values('Name', ascending = False)
pok.sort_values(['Type 1','HP'], ascending = [1,0])
```

# Making changes in data

```python
pok['total'] = pok.iloc[:, 4:10].sum(axis=1)


pok.head()
```
